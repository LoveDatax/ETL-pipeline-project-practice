```python
def extract(csv_path):
    try:
        df = pd.read_csv(csv_path)
        print(f"✅ Successfully extracted {len(df)} rows from {csv_path}")
        return df
    except Exception as e:
        print(f"❌ Extraction failed: {e}")
        return None
```

```python
csv_path = 'Downloads/archive/winemag-data_first150k.csv'
raw_data = extract(csv_path)
raw_data.head()
```

```python
def transform(df):
    df = df[['country', 'description', 'points', 'price', 'region_1', 'region_2']]
    df.fillna({'price': df['price'].mean(),
               'region_1':0, 'region_2':0}, inplace = True)
    df['points'].astype(int)
    return df
```

```python
transformed_data = transform(raw_data)
transformed_data
```

```python
transformed_data.info()
```

```python
n = 100  # Number of rows you want
subset = transformed_data.head(n)  # Selects the first n rows
```

```python
def load(clean_data, clean_data_file_path):
    clean_data.to_csv(clean_data_file_path, index = False)
```

```python
load(subset, 'transformed_wine_data.csv')
```
